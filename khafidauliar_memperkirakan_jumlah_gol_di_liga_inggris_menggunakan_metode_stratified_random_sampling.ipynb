```python
# Membaca data dari file CSV
data = pd.read_csv('/content/drive/MyDrive/Sampling/dataset - 2020-09-24.csv')

# Hitung total gol per posisi dalam populasi
total_gol_per_posisi = data.groupby('Position')['Goals'].sum()

# Hitung persentase jumlah gol per posisi
presentase_gol_per_posisi = (total_gol_per_posisi / data['Goals'].sum()) * 100

# Urutkan DataFrame berdasarkan urutan posisi: Goalkeeper, Defender, Midfielder, Forward
urutan_posisi = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
df_gol_posisi = pd.DataFrame({
    'Position': total_gol_per_posisi.index,
    'Goal': total_gol_per_posisi.values,
    'Presentase (%)': presentase_gol_per_posisi.round(2).values
}).sort_values('Position', key=lambda x: x.map({pos: idx for idx, pos in enumerate(urutan_posisi)}))

# Tambahkan total di bawah DataFrame
total_gol = df_gol_posisi['Goal'].sum()
total_presentase = df_gol_posisi['Presentase (%)'].sum()
df_total = pd.DataFrame({'Position': ['Total'],
                         'Goal': [total_gol],
                         'Presentase (%)': [total_presentase]})

df_gol_posisi = pd.concat([df_gol_posisi, df_total], ignore_index=True)

print(df_gol_posisi)
```

```python
# Hitung ukuran total populasi (N)
N = len(data)

# Tentukan tingkat kepercayaan (misalnya, 95%)
confidence_level = 0.95

# Tentukan margin of error (misalnya, 5%)
margin_of_error = 0.05

# Estimasi proporsi atau persentase dalam populasi (misalnya, 0.2)
estimated_proportion = 0.2

# Hitung ukuran sampel minimum dengan rumus Cochran
sample_size = (stats.norm.ppf(1 - ((1 - confidence_level) / 2)) ** 2 * estimated_proportion * (1 - estimated_proportion)) / margin_of_error ** 2

# Pastikan ukuran sampel tidak melebihi jumlah total populasi (N)
sample_size = min(sample_size, N)

# Konversi ukuran sampel menjadi integer
sample_size = int(sample_size)

print("Ukuran Sampel Minimum yang direkomendasikan:", sample_size)
```

```python
# Menambahkan kolom "Jumlah Sample" berdasarkan presentase dari masing-masing posisi
N = len(data)
n_total = sample_size
df_gol_posisi['Jumlah Sample'] = (df_gol_posisi['Presentase (%)'] / 100 * n_total).astype(int)

print(df_gol_posisi)
```

```python
# Fungsi untuk menghitung total jumlah gol per posisi pemain dalam populasi keseluruhan
def hitung_total_gol_per_posisi(data):
    total_gol_per_posisi = data.groupby(['Position', 'Club'])['Goals'].sum()
    return total_gol_per_posisi

# Fungsi untuk menghitung persentase jumlah gol per posisi pemain terhadap jumlah gol keseluruhan
def hitung_presentase_gol_per_posisi(total_gol_per_posisi, total_gol_keseluruhan):
    presentase_gol_per_posisi = (total_gol_per_posisi / total_gol_keseluruhan) * 100
    return presentase_gol_per_posisi

# Fungsi untuk menghitung jumlah sample berdasarkan presentase masing-masing posisi
def hitung_jumlah_sample(presentase, n_total):
    jumlah_sample = (presentase / 100) * n_total
    return jumlah_sample

# Hitung total gol per posisi dalam populasi
total_gol_per_posisi = hitung_total_gol_per_posisi(data)

# Hitung jumlah gol keseluruhan
total_gol_keseluruhan = data['Goals'].sum()

# Hitung persentase jumlah gol per posisi
presentase_gol_per_posisi = hitung_presentase_gol_per_posisi(total_gol_per_posisi, total_gol_keseluruhan)

# Gabungkan total gol per posisi dan persentase gol per posisi menjadi satu DataFrame
df_gol_posisi = pd.DataFrame({'Position': total_gol_per_posisi.index.get_level_values(0),
                              'Club': total_gol_per_posisi.index.get_level_values(1),
                              'Goal': total_gol_per_posisi.values,
                              'Presentase (%)': presentase_gol_per_posisi.values})

# Format nilai presentase agar menampilkan dua angka di belakang koma
df_gol_posisi['Presentase (%)'] = df_gol_posisi['Presentase (%)'].round(2)

# Urutkan DataFrame berdasarkan urutan posisi: Goalkeeper, Defender, Midfielder, Forward
urutan_posisi = ['Goalkeeper', 'Defender', 'Midfielder', 'Forward']
df_gol_posisi = df_gol_posisi[df_gol_posisi['Position'].isin(urutan_posisi)]

# Tambahkan total di bawah DataFrame menggunakan pd.concat()
total_gol = df_gol_posisi['Goal'].sum()
total_presentase = df_gol_posisi['Presentase (%)'].sum()
df_total = pd.DataFrame({'Position': ['Total'],
                         'Club': [''],
                         'Goal': [total_gol],
                         'Presentase (%)': [total_presentase]})
df_gol_posisi = pd.concat([df_gol_posisi, df_total], ignore_index=True)

# Menambahkan kolom "Jumlah Sample" berdasarkan presentase dari masing-masing posisi
N = len(data)
n_total = sample_size
df_gol_posisi['Jumlah Sample'] = (df_gol_posisi['Presentase (%)'] / 100 * n_total).astype(int)

# Menyusun ulang kolom agar tampilan lebih sesuai
df_gol_posisi = df_gol_posisi[['Position', 'Club', 'Goal', 'Presentase (%)', 'Jumlah Sample']]

df_gol_posisi
```
